


#### Abstract:

This notebook presents an analysis of the csv "BTC-USD.csv" meaning the daily Bitcoin to US dollars closing prices for the year 2023, we are particularly interested in Q3 corresponding to days 182 to 273 inclusive.

This analysis will be performed using NumPy for computations and matplotlib for visualisation. We focused on descriptive statistics, visual trends, and the identification of extreme values withing Q3 2023.


We begin by importing the packages numpy and matplotlib (optional: scipy).

```python
import numpy as np
import matplotlib.pyplot as plt
```

Found the dataset in the task resources: Used np.loadtxt() to load

```python
dataset = np.loadtxt("BTC-USD.csv", delimiter=",", skiprows = 1, usecols=4)

print("Total number of observations:", dataset.size)
```

The dataset contains daily closing prices for the entirety of 2023, so we expect to have 365 observations.

Extracting Q3 2023 Data:

Task sheet says: "days 182–273 inclusive" for the Q3

```python
q3 = dataset[181:273]

print("Total number of observations (q3):", q3.size)
```

## Descriptive Statistics

The third quarter consists of 92 daily observations. We will now calculate some metrics.

* Arithmetic mean
* Min
* First quartile
* Median
* Third quartile
* Max
* Std
* Interquartile range

```python
mean = np.mean(q3)
min = np.min(q3)
q1value = np.percentile(q3,25)
median = np.median(q3)
q3value = np.percentile(q3,75)
max = np.max(q3)
std = np.std(q3)
iqr = q3value - q1value

print(f"================")
print(f"Arithmetic mean: {mean:.2f}")
print(f"================")
print(f"Min: {min:.2f}")
print(f"================")
print(f"First quartile: {q1value:.2f}")
print(f"================")
print(f"Median: {median:.2f}")
print(f"================")
print(f"Third quartile: {q3value:.2f}")
print(f"================")
print(f"Max: {max:.2f}")
print(f"================")
print(f"Std: {std:.2f}")
print(f"================")
print(f"Interquartile range: {iqr:.2f}")
print(f"================")
```



The average closing price of BTC during Q3 of 2023 was 28k USD

Furthermore, **median > mean** meaning we have a slight left skew.
The minimum observed price was 25k USD and the maximum reached 31,5k USD.

**31476.05 - 25162.65 = 6313.4** indicates a total fluctuation of 6313 USD during Q3 of 2023.

Looking at Q1, median and Q3 we can interpret that half othe prices were between 26k and 29.7k USD.
Additionally, the Interquartile range shows that the 50% of BTC prices varied within about 3500 USD.

On average, daily closing prices deviated by **std = 1827.04**

## Visualisation of Q3 Prices

To go further and understand the behaviour of BTC during Q3 of 2023, we can plot the daily closing prices

- The horizontal axis represents the day number of the year, with 182 being the 1st of july
- we will use red solid line segments for this plot

```python
days = np.arange(182,274)

plt.plot(days, q3, color = "red")
plt.title("BTC - USD in Q3 of 2023")
plt.xlabel("day of the year")
plt.ylabel("BTC Closing Price in USD)")
plt.show()
```

## Interpretation of the Q3 2023 BTC Price Trend

From the plot, we can clearly see that BTC was trading close to 30,000–31,000 USD at the beginning of Q3. Actually, we can see at the start that the quarter started on a high BTC slosing price.

As the days progress, the price begins to slowly decline. The most noticeable change happens around the middle of the quarter, where there is a sharp drop. The price falls from roughly 29,000 USD down to almost 26,000 USD. This drop of around 3,000 USD indicates a significant shift in the market.

Later in the quarter, we see that BTC reaches its lowest point at approximately 25,200 USD (around day 254). This confirms that the second half of Q3 was weaker compared to the beginning.

Towards the end of the quarter, the price shows some recovery, rising again towards the 27k USD range. But it does not return to its peak, it suggests that the market didn't regain its previous momentum by the end of Q3.

Overall, the plot shows that Q3 2023 was quite volatile, with a clear mid-quarter decline followed by a partial recovery.

## Identification of the Extremes

we will determine the specific days corresponding to lowest and highest prices in Q3 of 2023

to do this we'll use np.argmin() and np.argmax() which will return the positions of min and max values in the Q3 vector

```python
minprice = np.argmin(q3)
maxprice = np.argmax(q3)

minday = 182 + minprice
maxday = 182 + maxprice

print(f"Lowest price was on day {minday} ({min:.2f}).")
print(f"highest price was on day {maxday} ({max:.2f}).")
```

The highest price happend early in the quarter while the lowest price happened towards the end of the quarter. This aligns well with what we see with our red line in our plot.

The statistical calculations and graphical represnetation indicate a period of "downfall" that is followed at the end by a moderate recovery.
