## 3. Data Wrangling

```python
# ── Basic info ────────────────────────────────────────────────────────────────
print(df.dtypes)
print(f'\nNull values: {df.isnull().sum().sum()}')
print(f'Date range : {df.index.min().date()} → {df.index.max().date()}')
print(f'Total days : {len(df)}')
```

```python
# ── Feature engineering ───────────────────────────────────────────────────────
df['Month']        = df.index.month
df['Month_Name']   = df.index.strftime('%B')
df['Quarter']      = df.index.quarter
df['Day_of_Week']  = df.index.day_name()
df['Daily_Return'] = df['Close'].pct_change() * 100   # % change day-over-day
df['7d_MA']        = df['Close'].rolling(7).mean()    # 7-day moving average
df['30d_MA']       = df['Close'].rolling(30).mean()   # 30-day moving average

df.head(10)
```

## 4. Full-Year 2023 Overview

```python
# ── Descriptive statistics ────────────────────────────────────────────────────
print('=== Full-Year 2023 – BTC-USD Closing Price Stats ===')
print(df['Close'].describe().round(2))
print(f'\nYTD return : {((df["Close"].iloc[-1] / df["Close"].iloc[0]) - 1) * 100:.1f}%')
```

```python
# ── Full-year closing price chart ─────────────────────────────────────────────
fig, ax = plt.subplots()
ax.plot(df.index, df['Close'], color='#F7931A', linewidth=1.4, label='Daily Close')
ax.plot(df.index, df['7d_MA'],  color='steelblue',  linewidth=1.2, linestyle='--', label='7-day MA')
ax.plot(df.index, df['30d_MA'], color='seagreen', linewidth=1.2, linestyle='--', label='30-day MA')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.set_title('BTC-USD Daily Closing Price – Full Year 2023', fontsize=14, fontweight='bold')
ax.set_ylabel('Price (USD)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ── Monthly average closing price ─────────────────────────────────────────────
monthly_avg = df.groupby('Month_Name')['Close'].mean().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]).round(2)

fig, ax = plt.subplots()
bars = ax.bar(monthly_avg.index, monthly_avg.values,
              color=['#F7931A' if m in ['July','August','September'] else '#AAAAAA'
                     for m in monthly_avg.index])
ax.set_title('Average Monthly BTC-USD Closing Price – 2023\n(Q3 highlighted in orange)', fontsize=13)
ax.set_ylabel('Avg. Close Price (USD)')
ax.set_xticks(range(len(monthly_avg)))
ax.set_xticklabels(monthly_avg.index, rotation=45, ha='right')
for bar, val in zip(bars, monthly_avg.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 200,
            f'${val:,.0f}', ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# ── Quarterly summary ─────────────────────────────────────────────────────────
quarterly = df.groupby('Quarter')['Close'].agg(
    Min='min', Max='max', Mean='mean', Std='std'
).round(2)
quarterly.index = ['Q1', 'Q2', 'Q3', 'Q4']
print('=== 2023 Quarterly Stats (BTC-USD Closing Price) ===')
print(quarterly)
```

## 5. Q3 2023 Deep-Dive (July – September)

Q3 2023 saw BTC-USD trade in a **relatively tight range** after the June surge. August experienced a notable sell-off, while July opened at highs and September closed the quarter near \$27 k.

```python
# ── Isolate Q3 ────────────────────────────────────────────────────────────────
q3 = df[df['Quarter'] == 3].copy()
print(f'Q3 records : {len(q3)}')
print(f'Q3 open    : ${q3["Close"].iloc[0]:,.2f}')
print(f'Q3 close   : ${q3["Close"].iloc[-1]:,.2f}')
print(f'Q3 high    : ${q3["Close"].max():,.2f}  ({q3["Close"].idxmax().date()})')
print(f'Q3 low     : ${q3["Close"].min():,.2f}  ({q3["Close"].idxmin().date()})')
q3_return = (q3['Close'].iloc[-1] / q3['Close'].iloc[0] - 1) * 100
print(f'Q3 return  : {q3_return:.1f}%')
```

```python
# ── Q3 closing price chart with month-shading ─────────────────────────────────
fig, ax = plt.subplots()

# colour bands per month
month_colors = {'July': '#FFF3E0', 'August': '#E3F2FD', 'September': '#E8F5E9'}
for month_name, color in month_colors.items():
    month_data = q3[q3['Month_Name'] == month_name]
    if not month_data.empty:
        ax.axvspan(month_data.index[0], month_data.index[-1], alpha=0.4, color=color, label=month_name)

ax.plot(q3.index, q3['Close'], color='#F7931A', linewidth=1.6, label='Daily Close')
ax.plot(q3.index, q3['7d_MA'],  color='steelblue',  linewidth=1.2, linestyle='--', label='7-day MA')

# annotate high / low
high_date = q3['Close'].idxmax()
low_date  = q3['Close'].idxmin()
ax.annotate(f"High\n${q3['Close'].max():,.0f}",
            xy=(high_date, q3['Close'].max()),
            xytext=(high_date, q3['Close'].max() + 500),
            arrowprops=dict(arrowstyle='->', color='green'), color='green', fontsize=9, ha='center')
ax.annotate(f"Low\n${q3['Close'].min():,.0f}",
            xy=(low_date, q3['Close'].min()),
            xytext=(low_date, q3['Close'].min() - 1200),
            arrowprops=dict(arrowstyle='->', color='red'), color='red', fontsize=9, ha='center')

ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b'))
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=0))
plt.xticks(rotation=45)
ax.set_title('BTC-USD Daily Closing Price – Q3 2023 (July, August, September)', fontsize=13, fontweight='bold')
ax.set_ylabel('Price (USD)')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ── Monthly descriptive stats within Q3 ──────────────────────────────────────
q3_monthly = q3.groupby('Month_Name')['Close'].agg(
    Open='first', Close_End='last', Min='min', Max='max', Mean='mean', Std='std'
).reindex(['July', 'August', 'September']).round(2)
q3_monthly['Range'] = (q3_monthly['Max'] - q3_monthly['Min']).round(2)
q3_monthly['Return_%'] = ((q3_monthly['Close_End'] / q3_monthly['Open']) - 1).mul(100).round(2)
print('=== Q3 2023 Monthly Summary ===')
print(q3_monthly.to_string())
```
